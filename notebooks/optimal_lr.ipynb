```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

xs = [8, 16, 32, 64, 128, 256, 512, 1024]
ys = [0.1, 0.1, 0.1, 0.2, 0.2, 0.2, 0.2, 0.5]

#linear fit
def linear_fit(x, a, b):
    return np.exp(a * np.log(x) + b)

popt, pcov = curve_fit(linear_fit, xs, ys)
print("Optimal learning rate (linear fit):", linear_fit(1024, *popt))

plt.scatter(xs, ys, marker="s")
plt.plot(xs, linear_fit(np.array(xs), *popt), label="Linear fit", color="red")

plt.xscale("log")
plt.yscale("log")
plt.xticks(xs, [f"{x}" for x in xs])
plt.xticks([], minor=True);
plt.xlabel("Batch size")
plt.ylabel("Optimal LR")
```
